# Array Operations

```
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
print(a1)
print(a2)
```

```
# scalar operations
print(a1*2)
print(a1+2)
print(a1-2)
print(a1/2)
print(a1**2)
```

```
# relational
print(a2>15)
print(a2==15)
```

```
# vector operations
print(a1*a2)
print(a1+a2)
print(a1-a2)
print(a1/a2)
print(a1**a2)
```

# array functions





```
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
print(a1)
print(a2)
```

```
#max/min/sum/prod
print(np.max(a1))
print(np.min(a1))
print(np.sum(a1)) # sum of all
print(np.prod(a1)) # product of all
```

```
np.max(a1,axis=0) # row wise # 0 means coloumns 1 means row
```

```
#mean/median/std/var
print(np.mean(a1))
print(np.median(a1))
print(np.std(a1))
print(np.var(a1))
```

```
# dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)  # 3*4 * 4*3 = 3*3
np.dot(a2,a3)
```

```
#log and exponent
np.log(a1)
```

```
print(np.round(np.random.random((2,3))*100)) # round off to nearest integer
print(np.floor(np.random.random((2,3))*100)) # previous integer
print(np.ceil(np.random.random((2,3))*100)) # next integer
```

# Stacking
