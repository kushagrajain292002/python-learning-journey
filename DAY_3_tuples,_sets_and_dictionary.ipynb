# Tuples

```
# create a tuple with a single item
t2 = ('a',)
print(t2)
# homogenous tuple
t1 = (1,2,3,4)
print(t1)
```

```
# indexing
t3 = ('a','b','c','d')
print(t3[0])
print(t3[-1])
print(t3[::-1])
```

```
# editing tuples
t3 = (1,2,3,4)
print(t3)
t3[0] = 10
print(t3)
# tuples are immutable
```

```
# adding items
# you cannot add items in tuples just like lists
```

```
# deleting
# you can delete whole not a part of it
```

```
# operations on tuples
# + and *
t1 = (1,2,3)
t2 = (4,5,6)
print(t1+t2)
print(t1*3)

# membership
print(1 in t1)

# iteration
for i in t1:
    print(i)
```

```
# tuples function
# len/sum/min/max
t1 = (1,2,3,4)
print(len(t1))
print(sum(t1))
print(min(t1))
print(max(t1))
print(sorted(t1))
```

```
# count
print(t1.count(2))
print(t1.index(2)) # tells the position
```

```
# special syntax
# tuples unpacking
a,b,c = (1,2,3)
print(a,b,c)
```

```
a=1
b=2
a,b=b,a
print(a,b)
```

```
a,b,*others = (1,2,3,4,5,6)
print(a,b,others)
print(others)
print(type(others))
```

```
# zipping tuples
l1 = [1,2,3]
l2 = [4,5,6]
print(tuple(zip(l1,l2)))
print(list(zip(l1,l2)))
```

# Sets

```
# unordered,mutable,no duplicates,can't contain mutuable data types
```

```
# creating sets
s1 = {1,2,3,4}
print(type(s1))
print(s1)
s4 = set([1,2,3,4])
print(s4)
s5 = set((1,2,3,4))
print(s5)
```

```
s1 = {1,2,3}
s2 = {3,2,1}
print(s1==s2)
```

```
# accessing items
s1 = {1,2,3,4}
print(s1[0])
print(s1[0] = 100)
# means no indexing and slicing
```

```
# adding and deleting items
s = {1,2,3,4}
s.add(5)
print(s)

```

```
s = {1,2,3,4}
s.remove(4)
print(s)
```

```
# update
s.update([6,7,8])
print(s)
```

```
s = {1,2,3,4}
del (s)
print(s)
```

```
s = {1,2,3,4,5}
s.discard(4)
print(s)
```

```
s.pop()
# delete any random item
```

```
s.clear()
print(s)
```

```
# operations on sets
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
# union
print(s1 | s2)
print(s1.union(s2))
# intersection
print(s1 & s2)
print(s1.intersection(s2))
#difference
print(s1 - s2)
print(s1.difference(s2))
# symmetric difference
print(s1^s2)
print(s1.symmetric_difference(s2))
```

```
#functions
#len/sum/min/max/sorted
s = {1,2,3,4}
print(len(s))
print(sum(s))
print(min(s))
print(max(s))
print(sorted((s),reverse=True))
```

```
#isdisjoint/issubset/issuperset
s1 = {1,2,3,4}
s2 ={5,6,8,9}

print(s1.isdisjoint(s2))
print(s1.issubset(s2))
print(s1.issuperset(s2))
```

```
# frozenset
# its just a immutable version of python set
# same differences between lists and tuples
fs = frozenset([1,2,3,4])
print(fs)
```

```
fs = frozenset([1,2,frozenset([3,4])])
fs
# 2d frozenset are possible
```

```
#set comprehension
{i for i in range(1,10)}
```

```
{s**2 for s in range(1,20) if s>5}
```

# Dictionary

```
i = {'name':'kushagra','age':22,'gender':'male'}
i
```

```
# accessing items
i = {'name':'kushagra','age':22,'gender':'male'}
i['name']
```

```
# adding key value
i = {'name':'kushagra','age':22,'gender':'male'}
i['height'] = 6
i['weight'] = 72
i
```

```
# remove key value
# pop,popitem,del,clear
```

```
i = {'name':'kushagra','age':22,'gender':'male'}
i.pop('age')
print(i)
```

```
i = {'name':'kushagra','age':22,'gender':'male'}
i.popitem()
print(i)
#remove last item



```

```
# editing key value pair
i = {'name':'kushagra','age':22,'gender':'male'}
i['name'] = 'chotu'
i
```

```
# dictionary operation
# membership and iteration
i = {'name':'kushagra','age':22,'gender':'male'}
'kushagra' in i # false
'name' in i
```

```
d = {'name':'kushagra','age':22,'gender':'male'}
for i in d:
  print(i,d[i])
```

```
# dictionary functions
#all same as len,sorted,max,min
```

```
#items
print(d.items())
#keys
print(d.keys())
#values
print(d.values())
```

```
# update
d1 = {'name':'kushagra','age':22,'gender':'male'}
d2 = {'age':6,'name':'chotu'}
d1.update(d2)
print(d1)
```

```
# DICTIONARY COMPREHENSION
# {key : value for vars in iterable}
# print 1st 10 numbers and thier squares
{ i:i**2 for i in range(1,11)}
```

```
# using existing dict
d = {'delhi':1000,'mumbai':2000,'banglore':3000}
{key:value*2 for (key,value) in d.items()}
```

```
#using zip
days = ['sun','mon','tue']
temp = [30,40,50]
{key:value for (key,value) in zip(days,temp)}
```

```
# using if
products = {'phone':10,'laptop':0,'charger':30,'tablet':0}
{key:value for (key,value) in products.items() if value>0}
```

```
# nested comprehension
# print tables of number from 2 to 4
{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
```
