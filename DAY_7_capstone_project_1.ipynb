```
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2017,1,1)
```

```
tesla = yf.download('TSLA',start ,end)
```

```
tesla.head()
```

```
ford = yf.download('F',start,end)
```

```
ford.head()
```

```
gm = yf.download('GM',start,end)
```

```
gm.head()
```

```

#  Create one figure and one set of axes
fig, ax = plt.subplots(figsize=(12, 8))

#  Plot all 'Open' lines on the same axes
tesla['Open'].plot(ax=ax, label='Tesla')
gm['Open'].plot(ax=ax, label='GM')
ford['Open'].plot(ax=ax, label='Ford')

#  Add labels and legend
ax.set_title('Opening Prices of Tesla, GM, and Ford (2023)')
ax.set_xlabel('Date')
ax.set_ylabel('Opening Price (USD)')
ax.legend()
plt.grid(True)
plt.show()
```
