# Adding items to the lists


```
# append
L = [1,2,3,4,5]
L.append(6)
print(L)
# we can append any data type
```

```
# extend
L = [1,2,3,4,5]
L.extend([6,7,8])
print(L)
# extend will break the items
```

```
# insert
# it add at your desire position
L = [1,2,3,4,5]
L.insert(1,100)
print(L)
```

# editing items on lists

```
L = [1,2,3,4,5]
L[-1] = 500
print(L)
```

```
L = [1,2,3,4,5]
L[1:4] = [200,300,400]
print(L)
```

# deleting items on list


```
L = [1,2,3,4,5]
del L[0]
print(L)
```

```
L = [1,2,3,4,5]
del L[1:3]
print(L)
```

```
# remove
L = [1,2,3,4,5]
L.remove(5)
# here you can delete by naming the value
print(L)
```

```
# pop
L = [1,2,3,4,5]

L.pop(0)

print(L)
```

```
# clear
L = [1,2,3,4,5]
L.clear()
print(L)
```

# Operations on lists
- arithmetic
- membership
- loop



```
# arithmetic (+,*)
L1 = [1,2,3,4]
L2 = [5,6,7,8]
print(L1+L2)
```

```
# membership
L1 = [1,2,3,4]
L2 = [1,2,3,4,[5,6]]

print(5 not in L1)
print(5 in L2)
```

```
# loops
L1 = [1,2,3,4]
L2 = [1,2,3,4,[5,6]]

for i in L1:
  print(i)

for i in L2:
  print(i)
```

# List functions


```
#len/min/max/sorted
L1 = [1,3,2,4]
print(len(L1))
print(min(L1))
print(max(L1))
print(sorted(L1))
```

```
# count
L = [1,2,3,1,13,1,1,3,4]
L.count(1)
```

```
# index
L = [1,2,3,1,13,1,1,3,4]
L.index(1)
```

```
L1 = [1,2,3,4]
L1.reverse()
print(L1)
```

```
# sort
L = [1,2,3,1,13,1,1,3,4]
L.sort()
print(L)
```

```
# copy
L1 = [1,2,3,4]
L2 = L1.copy()
print(L2)
```

# List comprehension

```
#newlist = [expression 'for' item 'in' iterable 'if' condition == 'True']
```

```
#Add 1 to 10 numbers to a list
L = []
for i in range(1,11):
  L.append(i)
print(L)
```

```
[i for i in range(1,11)]
```

```
v = [2,3,4]
s = -3
# [-6,-9,-12]

x = []
for i in v:
  x.append(i*s)
print(x)
print(v*s)
```

```
# add squares
L = [1,2,3,4]
[i**2 for i in L]
```

```
# print all the numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,51) if i%5 == 0]
```

```
# find languages which start with letter p
languages = ['java','python','php','c','javascript']
[i for i in languages if i.startswith('p')]
```

```
# nested if with list comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']
# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[i for i in my_fruits if i in basket if i.startswith('a')]
```

```
# print a (3,3) matrix using list comprehension -> nested list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```
# cartesian products -> list comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]
[i*j for i in L1 for j in L2]
```

# 2 ways to traverse a list





```
# itemwise
L = [1,2,3,4]
for i in L:
  print(i)
```

```
# indexwise
L = [1,2,3,4]
for i in range(0,len(L)):
  print(L[i])
```

```
# write a program to add items of 2 lists indexwise
L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))
[i+j for i,j in zip(L1,L2)]
```
